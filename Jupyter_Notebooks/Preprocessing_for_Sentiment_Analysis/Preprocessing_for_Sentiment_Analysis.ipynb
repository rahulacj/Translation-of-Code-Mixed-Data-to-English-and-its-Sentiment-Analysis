```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
```

```python
model_ckpt = "rahulacj/mbart-large-cc25-finetuned-hi-to-en-v2"
```

```python
my_model = AutoModelForSeq2SeqLM.from_pretrained(model_ckpt)
```

```python
my_tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
```

```python
if "mbart" in model_ckpt:
    my_tokenizer.src_lang = "hi_IN"
```
